## A1: Data Sets

There were two source data sets utilized for the Tableau project. First, the churn_clean CSV dataset that was provided with the coursework. Second, an American Community Survey on Types of Internet Subscriptions by Selected Characteristics found on the United States Census Bureau data platform(U.S. Census Bureau, 2022). The churn_clean and ACSST1Y2022.S2802 CSVs were submitted along with the project. Below is a Python script implemented to manipulate and combine the datasets for usage within Tableau.

```python
import pandas as pd 
import numpy as np
```

```python
churn = pd.read_csv('C:/Users/alexa/WGU/D210/Project/churn_clean.csv', index_col=0)

pd.set_option("display.max_columns", None)
```

```python
report_churn = churn[["State","Age","Churn","InternetService"]].copy()
```

```python
report_churn.loc[report_churn['Age'] <65 ,'Age_Group'] = '18 to 64 years'
report_churn.loc[report_churn['Age'] >=65 ,'Age_Group'] = '65 years and over'
report_churn['Internet'] = np.where(report_churn['InternetService'] == 'None', 'No', 'Yes')
report_churn = report_churn.replace('Yes', 1)
report_churn = report_churn.replace('No', 0)
report_churn['National'] = 'National'
report_churn
```

```python
national = pd.read_csv('C:/Users/alexa/WGU/D210/Project/ACSST1Y2022.S2802-2023-10-09T071926.csv', index_col=0)
```

```python
national = national[["United States!!With a computer!!Percent  Broadband Internet Subscription!!Estimate"]].copy()
```

```python
national = national.iloc[0:5].copy()
```

```python
national = national.dropna()
national.rename(columns={"United States!!With a computer!!Percent  Broadband Internet Subscription!!Estimate":"National_P"}, inplace=True)
```

```python
national.index.names = ['Groups']
national.reset_index(inplace=True)
national['National'] = 'National'
```

```python
national = national.pivot(index='National', columns = 'Groups', values='National_P')
national
```

```python
national = national[["18 to 64 years","65 years and over","Total population in households"]].copy()
national.reset_index(inplace=True)
```

```python
national = national.merge(report_churn, how='inner', left_on='National', right_on='National')
national
```

```python
national.loc[national['Age_Group'] == '18 to 64 years' ,'Age_GroupP'] = national['18 to 64 years']
national.loc[national['Age_Group'] == '65 years and over' ,'Age_GroupP'] = national['65 years and over']
national
```

```python
national.to_csv('C:/Users/alexa/WGU/D210/Project/National_Prepared_Data.csv', index=False)
```

## A3: Navigation Instructions

The interactive dashboard project is composed of a story with two story points, which is a minimalist way of displaying two dashboards created. Selecting National/State AVG on the top center loads the first dashboard which includes two filters on the top right for Customer Churn and State selections. The National/State AVG dashboard has four data representations: 

- Customer Nation Internet AVG on the top which highlights the current customer percentage with an internet subscription versus the national internet average and the variance to that national average. 
- Customer State Internet AVG shows the customer percentage of the current selection of the state filter with an internet subscription versus the national internet average and the variance to that national average.
- States Ranked by Internet AVG on the bottom left ranks all states in descending order by customer internet subscription versus the national reference line. 
- Customer Counts by State on the bottom right is a map with a visual size representation of customers within each state. 

The story point selection on the top center of Age Group AVG displays the relevant dashboard for Age groups with two filters for Age Group and State. The age group filter is a single selection dropdown that interacts with the state data presentations. This dashboard is composed of three data presentations:

- Customer Nation Internet Age AVG on the top highlights the current customer percentage with an internet subscription versus the national internet average and the variance to that national average. 
    - The data presentation and metrics are broken down into the two age groups of 18 to 64 years and 65 years & older.
- Customer State Internet Age AVG in the middle is the current selection of the state filter with an internet subscription versus the national internet age group average and the variance to that national age group average.

I made an attempt to minimize the use of color wherever possible and use grayscale in the dashboards.
